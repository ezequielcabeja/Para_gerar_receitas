```python
import requests
import random
import textwrap
from IPython.display import display, Markdown
import google.generativeai as palm
import os

# --- Funções auxiliares ---
def to_markdown(text):
    text = text.replace('•', ' *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

def exibir_historico(chat_history):
  for message in chat_history:
    display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
    print('-------------------------------------------')

# --- Funções para cada funcionalidade ---

def gerar_receita():
    """
    Gera uma receita completa usando o Gemini.
    """
    while True:
      user_query = input("Digite o nome da receita que você deseja (ou 'sair' para voltar ao menu): ")
      if user_query.lower() == 'sair':
        break
      recipe = get_recipe(user_query)
      print("\n## Receita para", user_query, ":\n")
      print(recipe)
      print("-" * 30)

def jogar_adivinhacao():
    """
    Joga um jogo de adivinhação de palavras com o usuário.
    """
    url = 'https://raw.githubusercontent.com/guilhermeonrails/api-imersao-ia/main/words.json'
    resposta = requests.get(url)
    data = resposta.json()
    valor_secreto = random.choice(data)
    palavra_secreta = valor_secreto['palavra']
    dica = valor_secreto['dica']
    print(f'A palavra secreta tem {len(palavra_secreta)} letras')
    print (f'A dica é -> {dica}')
    chute = input('O que você acha que é? R: ')
    if chute == palavra_secreta:
        print ('Acertou')
    else:
        print(f'Errou... a palavra secreta é ->{palavra_secreta}')

def conversar_ia():
    """
    Oferece sugestões de conteúdo sobre IA e conversa sobre o tema.
    """
    model = palm.GenerativeModel(model_name='gemini-pro')
    chat = model.start_chat(history=[])
    prompt = input('Esperando prompt (digite "fim" para terminar): ')
    while prompt != 'fim':
      response = chat.send_message(prompt)
      print('Resposta: ', response.text, "\n")
      prompt = input('Esperando prompt (digite "fim" para terminar): ')
    exibir_historico(chat.history)


# --- Função para obter a receita do Gemini ---
def get_recipe(query):
    """
    Gera uma receita completa usando o Gemini.

    Args:
        query: A consulta do usuário sobre a receita desejada.

    Returns:
        A receita gerada pelo Gemini, formatada como uma string.
    """
    prompt = f"""Você é um chef especialista em culinária.

    Um usuário solicitou a receita para: {query}.

    Por favor, forneça uma receita completa e detalhada, incluindo:

    * Título da receita
    * Ingredientes
    * Instruções passo-a-passo, numeradas para facilitar o acompanhamento
    * Tempo de preparo
    * Tempo de cozimento
    * Rendimento
    * Dicas (opcional)
    """
    response = palm.generate_text(
        model='gemini-pro',
        prompt=prompt,
        temperature=0.7,
        max_output_tokens=500,
    )
    return response.text

# --- Configuração do Gemini ---
gemini_api_key = os.environ.get('GEMINI_API_KEY')
if not gemini_api_key:
  gemini_api_key = input("Por favor, insira sua chave de API do Gemini: ")
palm.configure(api_key=gemini_api_key)

# --- Loop principal ---
while True:
    print("\nEscolha uma opção:")
    print("1 - Gerar Receita")
    print("2 - Jogar Adivinhação")
    print("3 - Conversar sobre IA")
    print("4 - Sair")
    escolha = input("Digite o número da opção desejada: ")

    if escolha == '1':
        gerar_receita()
    elif escolha == '2':
        jogar_adivinhacao()
    elif escolha == '3':
        conversar_ia()
    elif escolha == '4':
        break
    else:
        print("Opção inválida. Tente novamente.")
```
